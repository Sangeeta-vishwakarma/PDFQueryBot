```python
from PyPDF2 import PdfReader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.llms import HuggingFaceEndpoint
from langchain_huggingface import ChatHuggingFace,HuggingFacePipeline
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
import os
```

```python
from google.colab import userdata
os.environ["HUGGINGFACEHUB_ACCESS_TOKEN"] = "YOUR API TOKEN KEY"
```

```python
# Step 1: Load PDF
pdfreader = PdfReader("YOUR_PDF_FILE.pdf")
```

```python
# chunk pdf into pages and pages into text
raw_text = ""
for page in pdfreader.pages:
    content = page.extract_text()
    if content:
        raw_text += content
```

```python
# Step 2: Split Text
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    length_function = len,
    separators = ["\n\n", "\n", " ", ""]
)
texts = text_splitter.split_text(raw_text)
```

```python
raw_text
```

```python
# Step 3: Embedding Model (open source)
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# Step 4: Vector Store (Chroma in-memory)
vectorstore = Chroma.from_texts(texts, embedding)
```

```python
# Step 5: Retriever
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 4})
```

```python
from huggingface_hub import login
login(token=os.getenv('HUGGINGFACE_ACCESS_TOKEN'))
```

```python
llm = HuggingFacePipeline.from_model_id(
    model_id="google/gemma-3-1b-it",
    task = "text-generation",
    pipeline_kwargs=dict(
        temperature= 0.7,
        max_new_tokens=100
    ),
)
```

```python
model = ChatHuggingFace(llm=llm)
```

```python
# Step 7: Prompt + Chain
prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
    Answer the question based on the context below.
    If the answer is not in the context, just say "I don't know."

    Context: {context}
    Question: {question}
    """
)
```

```python
# define parser
parser = StrOutputParser()
```

```python
# define chain using all components
chain = prompt | model | parser
```

```python
# Step 8: Query Function
def ask_question(query):
    docs = retriever.get_relevant_documents(query)
    context = "\n\n".join([doc.page_content for doc in docs])
    return chain.invoke({"context": context, "question": query})
```

```python
ask_question("explain bias variance trade off")
```

```python
ask_question("what is the difference between supervised and unsupervised learning")
```
